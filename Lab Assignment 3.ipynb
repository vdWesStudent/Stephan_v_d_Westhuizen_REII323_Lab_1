```
#Task 7
import csv
inputFile = open('duplicate.csv','r')

outputFile = open('noDuplicate.csv','w')

listLines = []

for line in inputFile:

    if line in listLines:
        continue

    else:
        outputFile.write(line)
        listLines.append(line)

outputFile.close()
inputFile.close()
print("Duplicates removed and saved in noDuplicates.csv")
```

```
#Task 8
import csv
input = open('messy.csv', 'r')
output = open('notMessy.csv', 'w')
writer = csv.writer(output)
for row in csv.reader(input,delimiter = ','):
    if any(field.strip() for field in row):
        writer.writerow(row)
input.close()
output.close() 
print("Whitespaces removed")
```
