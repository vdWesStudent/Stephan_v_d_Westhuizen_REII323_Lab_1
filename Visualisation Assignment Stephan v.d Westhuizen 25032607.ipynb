```python
import pandas as pd
from matplotlib import pyplot as plt
import sklearn.datasets

def get_iris_df():
 ds = sklearn.datasets.load_iris()
 df = pd.DataFrame(ds['data'],
 columns = ds['feature_names'])
 code_species_map = dict(zip(
 range(3), ds['target_names']))
 df['species'] = [code_species_map[c]
 for c in ds['target']]
 return df

df = get_iris_df()

sums_by_species = df.groupby('species').sum()
var = 'petal width (cm)'
sums_by_species[var].plot(kind='pie', fontsize=15)
plt.ylabel(var, horizontalalignment='left')
plt.title('Breakdown for ' + var, fontsize=20)
plt.savefig('plot1_PIE.jpg')
plt.show()
print("From this pie char, one can see that the virginica species has the widest petals")
```

```python
sums_by_species = df.groupby('species').sum()
var = 'petal width (cm)'
sums_by_species[var].plot(kind='bar', fontsize=15,
rot=30)
plt.title('Breakdown for ' + var, fontsize=20)
plt.savefig('plot2_BAR.jpg')
plt.show()
print("This bar graph also indicates the virginica species has the widest petals.")
```

```python
for spec in df['species'].unique():
 forspec = df[df['species']==spec]
 forspec['petal length (cm)'].plot(
 kind='hist', alpha=0.6, label=spec)
plt.legend(loc='upper right')
plt.suptitle('Petal Length by Species')
plt.savefig('plot3_HISTO.jpg')
plt.show()
print("This histogram plots the different species' petal lengths, it clearly indicates that most of the virginica species' petals are longer than the other two species.")
```

```python
col = 'sepal length (cm)'
df['ind'] = pd.Series(df.index).apply(lambda i: i% 50)
df.pivot('ind','species')[col].plot(kind='box')
plt.title("Distribution of sepal length of different species")
plt.savefig('plot4_BOX.jpg')
plt.show()
print("This figure depicts the distribution of the sepal length of each species, one can clearly see where there are extreme values and where the mean of the data is.")
```

```python
colors = ["r", "g", "b"]
markers= [".", "*", "x"]
fig, ax = plt.subplots(1, 1)
plt.title("Petal Length vs Width by Species")
for i, spec in enumerate(df['species'].unique() ):
 ddf = df[df['species']==spec]
 ddf.plot(kind="scatter",
 x="petal width (cm)", y="petal length (cm)",
 alpha=0.5, s=10*(i+1), ax=ax,
 color=colors[i], marker=markers[i], label=spec)
plt.legend()
plt.savefig('plot5_SCAT.jpg')
plt.show()
print("The figure again indicates how the data is distributed, one can clearly see that setosa species has small petals.")
```

```python
df.plot(kind="hexbin",
 x="petal width (cm)", y="petal length (cm)")
plt.title("Heatmap of petal length vs height")
plt.savefig('plot6_HEAT.jpg')
plt.show()
print("The figure again indicates how the data is distributed, the colour scale on the right indicates how dense the data points are, ie how many data points overlap ")
```
